```python
%matplotlib inline

import cv2 # The OpenCV library; install using `pip install opencv-contrib-python`
import numpy as np # Helpful when working with arrays; install using `pip install numpy`
from matplotlib import pyplot as plot # Good for graphing; install using `pip install matplotlib`
from matplotlib import image as image
import easygui # An easy-to-use file-picker; pip install easygui
```

## Display video

```python
# Video capture from a file:
Video = cv2.VideoCapture("Zorro.mp4")
(check, I) = Video.read()

while check:
    cv2.imshow("image", I)
    
    # This delays for 1ms while waiting for a key to be pressed
    key = cv2.waitKey(1)

    # if the 'q' keqy is pressed, quit:
    if key == ord("q"):
         break

    # Next Frame:
    (check, I) = Video.read()

Video.release()
cv2.destroyAllWindows()
```

## Read in image (from video)

```python
# Video capture from a file:
frame_no = 24
Video = cv2.VideoCapture("../Films/Zorro.mp4")
Video.set(1,frame_no); # Where frame_no is the frame you want
(check, I) = Video.read()
cv2.imshow("image", I)

# This delays for 1ms while waiting for a key to be pressed
key = cv2.waitKey(10000)

Video.release()
cv2.destroyAllWindows()
```

## Convert to YUV

```python
YUV = cv2.cvtColor(I, cv2.COLOR_BGR2YUV)
#print("image")
#print(YUV)
#print("row")
#print(YUV[0,:])
#print("column")
#print(YUV[:,0])
#print("Y")
#print(YUV[0][0][0])
```

## Extract Y

```python
y = YUV[:,:,0]
print(y)
```

## Convert to Grayscale

```python
G = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
```

## Display Luminance and Grayscale using OpenCV

```python
cv2.imshow("Luminance", y)
cv2.imshow("Gray", G)
key = cv2.waitKey(10000)
cv2.destroyAllWindows()
```

## Showing both side by side using subplot

```python
fig, axs = plot.subplots(1, 2)
axs[0].set_title('Luminance')
axs[1].set_title('Intensity')
axs[0].imshow(y, cmap='gray')
axs[1].imshow(G, cmap='gray')
```
